# Example Floor Plan Ingestion and Visualization

CITAM provides facility management features to allow users to use real facilities in their simulations. Facilities are saved in the system's application data directory. A facility is determined by a name and list of floors. Each floor includes a map and a navigation network. All facility data is serialized and saved in JSON.

We start by importing the citam library:

```python
import citam
```

## Ingest Floorplan

To Ingest a new floorplan, you need to provide an SVG file and an optional CSV file. The CSV file will contain additinal information about the spaces defined in the SVG file. The `examples` folder has example floor plans to get started.

We will ingest the floorplan defined by the following SVG and CSV files: `examples/basic_example/TF1.svg` and `examples/basic_example/TF1.svg`. 

A facility and a floor names are required for the ingestion process to work. If the facility name provided does not exist yet, it will be created. Otherwise, a new floor will be added to the existing facility. Please note that if the floor name already exists, the process will fail because each floor needs a unique name. The inget_floorplan function takes additional parameters as defined in the documentation.

```python
citam.ingest_floorplan(facility="my_facility", floor="my_floor", svg="basic_example/TF1.svg", csv="basic_example/TF1.csv", verbose=1)
```

## List Facilities

To view a list of facilities that have already been ingested by CITAM, use the `list_facilities()` function:

```python
citam.list_facilities()
```

A summary table of each facility can be provided 

```python
citam.show_facility_summary("TF1")
```

## Load and Show Floorplans

A facility can have more than one floor. To load the floors, you can use the `load_floorplans()` function which will read the map files and create a list of `Floorplan` objects, one for each floor.

```python
floorplans = citam.load_floorplans(["0"], "TF1")
```

```python
floorplans[0].show()
```

```python
floorplans[0].show(include_doors=True)
```

`show_floorplan() is a convenience function that can load and display a floorplan in one step. It takes the facility and floor names as paramter.

```python
citam.show_floorplan("TF1", "0")
```

## Visualize Navigation Networks

```python
citam.show_navigation_network("TF1", "0")
```
